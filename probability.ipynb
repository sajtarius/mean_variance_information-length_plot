# Probability Density Function 
<p> The probaility density functions ($p(x,t)$) versus the spatial location ($x$) for different times ($t$) are plotted as shown in the following. In this plotting, the marginal probability density function is expressed as $p(x,t) = \sqrt{\frac{\beta_T}{\pi}} e^{-\beta_T \left[ x - x_0 e^{-\gamma t} \right]^2}$ such that $\beta_T = \frac{\beta_0 \beta}{\beta_0 + \beta e^{-2\gamma t}}$ and $\beta = \frac{\gamma}{2D\left[ 1-e^{-2\gamma t} \right]}$. The values for $\beta_0$, $x_0$, $\gamma$, and $D$ are constants that set to $1$ for this example. </p>

<p> For comparison, plots for different values of $t$ are shown. The values of $t$ are $t_1=0, t_2=0.01, t_3=0.1, t_4=1.0, t_5=10.0, t_6=100.0$ </p>

<p> <b> NOTE: </b> It is advisible to obtain $\beta_T$ by 'inversing' the result of $\frac{1}{2\beta_T} = \frac{D\left( 1-e^{-2\gamma t} \right)}{\gamma} + \frac{e^{-2\gamma t}}{2\beta_0}$. This is to avoid the 'undefined error' when $t = 0$. </p>
<p> <b> NOTE: </b> The mean will be shifted from the initial position, $x_0$, because of [mean function], $x_0 e^{-\gamma t}$. It will decay to ZERO due to exponential time decay. </p>

```python
#define the marginal probability density function
def pro(x, g, D, t, b0, x0):
    x = (D*(1-np.exp(-2*g*t)))/(g) + (np.exp(-2*g*t))/(2*b0)
    bt = 1/(2*x) #inversing the result
    return np.sqrt(bt/np.pi)*np.exp(-bt*(i-x0*np.exp(-g*t))**2)

#x-axis values and empty set for data point storing 
x = np.linspace(-5,5,100)
ans1=[]
ans2=[]
ans3=[]
ans4=[]
ans5=[]
ans6=[]

#parameter values 
g=1
D=1
b0=1
x0=0
t1=0
t2=0.01
t3=0.1
t4=1.0
t5=10.0
t6=100.0

#calculation of the data point
for i in x:
    ans11 = pro(x, g, D, t1, b0, x0)
    ans22 = pro(x, g, D, t2, b0, x0)
    ans33 = pro(x, g, D, t3, b0, x0)
    ans44 = pro(x, g, D, t4, b0, x0)
    ans55 = pro(x, g, D, t5, b0, x0)
    ans66 = pro(x, g, D, t6, b0, x0)
    ans1.append(ans11)
    ans2.append(ans22)
    ans3.append(ans33)
    ans4.append(ans44)
    ans5.append(ans55)
    ans6.append(ans66)
    
#plotting
a1,=plt.plot(x, ans1, label='t=0')
a2,=plt.plot(x, ans2, label='t=0.01')
a3,=plt.plot(x, ans3, label='t=0.1')
a4,=plt.plot(x, ans4, label='t=1.0')
a5,=plt.plot(x, ans5, label='t=10.0')
a6,=plt.plot(x, ans6, label='t=100.0')
plt.xlabel('x', fontsize=14)
plt.ylabel('probabilty density function', fontsize=14)
plt.legend(handles=[a1,a2,a3,a4,a5,a6], loc='upper right',fontsize=8)
plt.show()
```
